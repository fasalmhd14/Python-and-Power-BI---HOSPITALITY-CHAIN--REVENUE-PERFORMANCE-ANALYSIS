```python
#Exercise-1. In aggregate bookings find columns that have null values. Fill these null values with whatever you think is the appropriate subtitute
#(possible ways is to use mean or median)
```

```python
df_agg_bookings.isnull().sum()
```

```python
df_agg_bookings.capacity.describe()
```

```python
df_agg_bookings[df_agg_bookings.capacity.isnull()]
```

```python
df_agg_bookings = df_agg_bookings.fillna({
    'capacity' : df_agg_bookings.capacity.mean()
})
df_agg_bookings
```

```python
df_agg_bookings.isnull().sum() #hence NA value from capacity column is removed.
```

```python
#Exercise-2. In aggregate bookings find out records that have successful_bookings value greater than capacity. 
#Filter those records
```

```python
df_agg_bookings
```

```python
df_agg_bookings = df_agg_bookings[df_agg_bookings.successful_bookings <= df_agg_bookings.capacity]
df_agg_bookings
```

```python
# DATA TRANSFORMATION
```

```python
df_agg_bookings.head()
```

```python
df_agg_bookings["occ_pct"] = df_agg_bookings["successful_bookings"] / df_agg_bookings["capacity"]
df_agg_bookings
```

```python
df_agg_bookings["occ_pct"] = df_agg_bookings["occ_pct"].apply(lambda x: round(x*100,2))
df_agg_bookings.head()
                                                            
                                                              
```

```python
# 1. What is an average occupancy rate in each of the room categories?
```

```python
df_agg_bookings.head()
```

```python
df_agg_bookings.groupby("room_category").occ_pct.mean().round(2)
```

```python
df_rooms
```

```python
df_avg_rc = pd.merge(df_agg_bookings.groupby("room_category").occ_pct.mean().round(2), df_rooms, left_on = "room_category", right_on = "room_id")
df_avg_rc = df_avg_rc.drop("room_id", axis = "columns")
df_avg_rc
```

```python
#2. Print average occupancy rate per city
```

```python
df_hotels.head()
```

```python
df_avg_occ_city = pd.merge(df_hotels, df_agg_bookings, on = "property_id")
df_avg_occ = df_avg_occ_city.groupby("city").occ_pct.mean().round(2)
df_avg_occ
```

```python
#In the month of June, what is the occupancy for different cities
```

```python
df_date.head()
```

```python
df_jun_occ_cit = pd.merge(df_agg_bookings, df_hotels, on = "property_id")
df_jun_occ_cit = pd.merge(df_jun_occ_cit, df_date, left_on = "check_in_date" , right_on = "date")
df_jun_occ_cit = df_jun_occ_cit[df_jun_occ_cit["mmm yy"] == "Jun 22"]
df_jun_occ_cit = df_jun_occ_cit.groupby("city").occ_pct.mean().round(2).sort_values()
df_jun_occ_cit
```
